```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import sys
import pandas as pd
from DDPG_Agent import DDPG
from task import Task

num_episodes = 500
target_position = np.array([0., 0., 100.])
task = Task(target_pos=target_position)
agent = DDPG(task) 

rewards = []
avg_rewards = []
position_x = []
position_y = []
position_z = []


for i_episode in range(1, num_episodes+1):
    state = agent.reset_episode() # start a new episode
    episode_reward = 0
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(action, reward, next_state, done)
        state = next_state
        
        
        episode_reward += reward
        if done:
            print("i_episode: {}, reward: {}, average _reward: {} \n".format(i_episode, np.round(episode_reward, decimals=2), np.mean(rewards[-10:])))
            break

    
    rewards.append(episode_reward)
    avg_rewards.append(np.mean(rewards[-10:]))
    
    position_x.append(task.sim.pose[0])
    position_y.append(task.sim.pose[1])
    position_z.append(task.sim.pose[2])
         
sys.stdout.flush()   
    
init_pose=np.array([0., 0., 0., 0., 0., 0.])
    
```

## Plotting Rewards & Positions


```python
plt.figure(figsize=(20,10))
plt.plot(rewards, label='rewards')
plt.plot(avg_rewards, label='avg_rewards')
plt.legend()
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(position_x, label='x')
plt.plot(position_y, label='y')
plt.plot(position_z, label='z')
plt.legend()
plt.xlabel('Episode')
plt.ylabel('Positions')
plt.show()
```
